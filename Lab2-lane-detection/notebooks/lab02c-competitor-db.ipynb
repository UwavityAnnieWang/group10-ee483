## The Duckiebot lane detector

Our company got information about the lane detector that the duckiebot uses.  
We can only run the code but we do not have the source code.  
Let's test the their lane detector.  


The docker container from Lab 1 should already be in the MM.  
We just need to sync the new available code to the MM. 

1. In the HostOS terminal, go to the `lab-devel` directory using the `cd` command.  
Double check that you are in the `lab-devel` folder with the `pwd` command. 
It should return: `repo_folder/lab-devel`

2. Running the dts devel run command



```python
dts devel run -H <MMname>.local -s -M --cmd bash -f 
```

3. Once the MM shell appears, built and activate the catkin workspace

```python
catkin_build
source /code/catkin_ws/devel/setup.bash
```

4. Running the competitor's code.  In the packages folder, there is a package called `comp_ld`. This package only has a launch file called `cmp_ld.launch`.  Let's run it. 

```python
roslaunch cmp_ld cmp_ld.launch
```

You should obtain the following result.  That means that the competitor's lane detector is running.
Leave this terminal running.


5. Observing their detector images.
Open a new terminal in your VM. It can be terminals in the VS Code (not the browser one).  
Run the following command.

```python
dts start_gui_tools ee483mm<number>
```

This command allows us to open a new terminal tab that can connect visuals to the notebook. 
The usual dts devel run attach cannot connect to the notebook monitor while start_gui_tools can.

6. Once the MM shell appears, run `rqt_image_view`

Explore the images that generated by the competitor's code. All images are in topics starting with `ee483mm<number>/line_detector_node/debug` and ending in `compressed`
- The image in `edges/compressed` shows the countours of the lanes
- The image in `maps/compressed` has the filtered image for white pixels
- The image in `segments/compressed` outpus the line segments drawn on. Your node should output similar to this.

You can use these images to compare with your lane detector.
